### Importing modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy 
import seaborn 
from sklearn.linear_model import LinearRegression

seaborn.set()

data = pd.read_csv('sample1.csv')

```

### Defining dependent and independent variables


```python

#Output
y = data['GPA']

#Input
x = data['SAT']
```


### Plotting the data

```python

plt.scatter(x,y)
plt.xlabel('SAT',fontsize=20)
plt.ylabel('SAT',fontsize=20)
plt.show()
```


### Regression

```python


#Converting into matrix(vector)
x_matrix = x.values.reshape(-1,1)
print(x_matrix.shape)

#regression
lreg = LinearRegression()

lreg.fit(x_matrix,y)

#R-Squared (predict/actual)
R_ = lreg.score(x_matrix,y)

#Getting coefficients
b1 = lreg.coef_

#Getting intercepts
b0 = lreg.intercept_
```

### Plotting results

```python

plt.scatter(x,y)

yhat =b0 + b1*x 
fig = plt.plot(x,yhat,lw=4,c='orange',label='Regression Line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```
