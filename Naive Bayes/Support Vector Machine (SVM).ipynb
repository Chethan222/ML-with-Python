### Support Vector Machines

* Works well for classifying higher dimensional data (higher features and dimensions)
* Finds higher-dimensional support vectors across which to devide
  the data (support vector->Hyper planes)
* Kernel trick to represent data in higher-dimensional spaces 
  to find hyperplanes that might not be apparent in lower dimensions
* Data may come from camera etc.
